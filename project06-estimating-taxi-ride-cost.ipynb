```python
train_df=pd.read_csv('/kaggle/input/competitions/mlp-term-1-2026-kaggle-assignment-1/train.csv')
test_df=pd.read_csv('/kaggle/input/competitions/mlp-term-1-2026-kaggle-assignment-1/test.csv')
sample_df=pd.read_csv('/kaggle/input/competitions/mlp-term-1-2026-kaggle-assignment-1/sample_submission.csv')

print(train_df.shape)
print(test_df.shape)
print(sample_df.shape)
```

```python
train_data=train_df.drop(columns=['VendorID','tpep_pickup_datetime', 'tpep_dropoff_datetime','RatecodeID', 'total_amount'], axis=1)
test_data=test_df.drop(columns=['id','VendorID','tpep_pickup_datetime', 'tpep_dropoff_datetime','RatecodeID'], axis=1)
y_data=train_df['total_amount']
print(train_data.shape)
print(test_data.shape)
```

```python
train_data.info()
```

```python
cat_cols=train_data.select_dtypes(include='object').columns.tolist()
num_cols=test_data.select_dtypes(include='number').columns.tolist()
print(num_cols)
print(cat_cols)
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_data, y_data, test_size=0.2, random_state=9)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
num_pipe=Pipeline(steps=([
    ('impute1', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())]))

cat_pipe=Pipeline(steps=([
    ('impute2', SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder(handle_unknown='ignore', drop='first'))
]))

transformer=([
    ('num',num_pipe, num_cols),
    ('cat',cat_pipe, cat_cols)
])
preprocess1=ColumnTransformer(transformer, remainder='passthrough')
preprocess1
```

```python
X_train_preprocess=preprocess1.fit_transform(X_train)
X_test_preprocess=preprocess1.transform(X_test)
# test_data_preprocess=preprocess1.transform(test_data)
```

```python
test_data_preprocess=preprocess1.transform(test_data)
```

```python
ohe_columns = preprocess1.named_transformers_['cat']\
.named_steps['encoder'].get_feature_names_out(cat_cols)

final_columns = num_cols + list(ohe_columns)

X_train = pd.DataFrame(X_train_preprocess, columns=final_columns)
X_test = pd.DataFrame(X_test_preprocess, columns=final_columns)
test_final=pd.DataFrame(test_data_preprocess, columns=final_columns)

models = {
    'LinReg': LinearRegression,
    'Ridge': Ridge,
    'Lasso': Lasso,
    'Huber': HuberRegressor,
    'Tree': DecisionTreeRegressor,
    'RF': RandomForestRegressor,
    'GBR': GradientBoostingRegressor,
    'HGBR': HistGradientBoostingRegressor,
    'ABR': AdaBoostRegressor,
    'xgb': XGBRegressor
}

for name, model in models.items():

    model_instance = model()

    print("="*60)
    print(f"Training : {name}")

    model_instance.fit(X_train, y_train)

    y_pred = model_instance.predict(X_test)

    R2Score = r2_score(y_test, y_pred)
    mse_score = mean_squared_error(y_test, y_pred)

    print(f"R2 Score : {R2Score}")
    print(f"MSE Score : {mse_score}")
    print()
```
