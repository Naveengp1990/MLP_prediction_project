```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.12,
    stratify=y,
    random_state=4002
)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
cat_cols=X_train.select_dtypes(include=['object']).columns.tolist()
num_cols=X_train.select_dtypes(exclude=['object']).columns.tolist()
num_cols
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MaxAbsScaler

text_col = "comment"
num_cols = X_train.drop(columns=[text_col]).columns

preprocessor = ColumnTransformer([
    ("text", TfidfVectorizer(
        max_features=10000,
        ngram_range=(1,1),
        stop_words='english',
        sublinear_tf=True,
        token_pattern=r'[a-zA-Z0-9]+'), text_col),

    ("num", MaxAbsScaler(), num_cols)
])
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from sklearn.ensemble import HistGradientBoostingClassifier

models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, n_jobs=-1),
    #"LinearSVC": LinearSVC(),
    "MultinomialNB": MultinomialNB(),
    #"SGDClassifier": SGDClassifier(loss="log_loss", max_iter=1000),
    #"RandomForest": RandomForestClassifier(n_estimators=200, n_jobs=-1),
    #"xgb_model": XGBClassifier(
    # n_estimators=300,
    # max_depth=6,
    # learning_rate=0.1,
    # subsample=0.8,
    # colsample_bytree=0.8,
    # tree_method="hist",
    # eval_metric="mlogloss",
    # n_jobs=-1),
    # "hgb_model": HistGradientBoostingClassifier(
    # max_depth=8,
    # learning_rate=0.08,
    # max_iter=300)

}

results = {}

for name, model in models.items():

    pipe = Pipeline([
        ("prep", preprocessor),
        ("sampler", RandomOverSampler(random_state=412)),
        ("model", model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    print("\n", "="*50)
    print(f"Model: {name}")
    print(classification_report(y_test, y_pred))

    results[name] = classification_report(y_test, y_pred, output_dict=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
```

```python
text_col = "comment"

#num_cols = [
    # "post_id","emoticon_1","emoticon_2","emoticon_3",
    # "upvote","downvote","if_1","if_2",
    # "race","religion","gender","disability",
    # "year","month","day","hour"]

preprocessor = ColumnTransformer([
    ("text", TfidfVectorizer(max_features=10000, ngram_range=(1,1)), text_col),
    ("num", MaxAbsScaler(), num_cols)
    
], remainder='passthrough')

#("num", StandardScaler(), num_cols)
```

```python
from imblearn.pipeline import Pipeline
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB, CategoricalNB
```

```python
# pipeline = Pipeline([
#     ("prep", preprocessor),
#     ("sampler", RandomOverSampler(random_state=42)),
#     ("model", LogisticRegression())
# ])
# pipeline
pipeline = Pipeline([
    ("prep", preprocessor),
    ("sampler", RandomOverSampler(random_state=412)),
    ("model", LogisticRegression(max_iter=1000, n_jobs=-1))
])
pipeline
```

```python
from sklearn.metrics import classification_report, confusion_matrix

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "model__C":[0.1,1,5],
    "prep__text__max_features":[10000,15000]
}

gs = GridSearchCV(pipeline, params, cv=3, scoring="f1_macro")
gs.fit(X_train, y_train)

print(gs.best_params_)
```
