```python
# ============================================================
# HOUSE PRICES - ADVANCED REGRESSION TECHNIQUES
# OPTIMIZED KAGGLE SOLUTION
# ============================================================

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from scipy.stats import skew
from scipy.special import boxcox1p
from scipy.stats import boxcox_normmax

from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder

from sklearn.linear_model import ElasticNet
from sklearn.ensemble import GradientBoostingRegressor

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# ============================================================
# LOAD DATA
# ============================================================

train_df = pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv')

test_df = pd.read_csv(
    '/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv'
)

# Save IDs
train_ids = train_df['Id']
test_ids = test_df['Id']

# ============================================================
# REMOVE OUTLIERS
# ============================================================

train_df = train_df.drop(
    train_df[
        (train_df['GrLivArea'] > 4000) &
        (train_df['SalePrice'] < 300000)
    ].index
)

# ============================================================
# TARGET TRANSFORMATION
# ============================================================

# Kaggle metric uses log
y = np.log1p(train_df['SalePrice'])

# ============================================================
# COMBINE TRAIN + TEST
# ============================================================

train_features = train_df.drop(['SalePrice'], axis=1)
all_data = pd.concat([train_features, test_df]).reset_index(drop=True)

# ============================================================
# FEATURE ENGINEERING
# ============================================================

def create_features(df):

    df = df.copy()

    # Total SF
    df['TotalSF'] = (
        df['TotalBsmtSF'] +
        df['1stFlrSF'] +
        df['2ndFlrSF']
    )

    # Total bathrooms
    df['TotalBath'] = (
        df['FullBath'] +
        0.5 * df['HalfBath'] +
        df['BsmtFullBath'] +
        0.5 * df['BsmtHalfBath']
    )

    # Total porch area
    df['TotalPorchSF'] = (
        df['OpenPorchSF'] +
        df['3SsnPorch'] +
        df['EnclosedPorch'] +
        df['ScreenPorch'] +
        df['WoodDeckSF']
    )

    # House age
    df['HouseAge'] = df['YrSold'] - df['YearBuilt']

    # Remodel age
    df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']

    # Has basement
    df['HasBsmt'] = (df['TotalBsmtSF'] > 0).astype(int)

    # Has garage
    df['HasGarage'] = (df['GarageArea'] > 0).astype(int)

    # Has fireplace
    df['HasFireplace'] = (df['Fireplaces'] > 0).astype(int)

    return df

all_data = create_features(all_data)

# ============================================================
# DROP VERY HIGH MISSING COLUMNS
# ============================================================

drop_cols = [
    'PoolQC',
    'MiscFeature',
    'Alley',
    'Fence'
]

all_data.drop(columns=drop_cols, inplace=True)

# ============================================================
# NUMERICAL & CATEGORICAL COLUMNS
# ============================================================

numeric_cols = all_data.select_dtypes(include=np.number).columns
categorical_cols = all_data.select_dtypes(exclude=np.number).columns

# ============================================================
# SIMPLE SKEWNESS HANDLING (STABLE)
# ============================================================

from scipy.stats import skew

numeric_cols = all_data.select_dtypes(include=np.number).columns

skewed_features = all_data[numeric_cols].apply(
    lambda x: skew(x.dropna())
)

skewed_features = skewed_features[
    abs(skewed_features) > 1
]

print(f"Skewed Features Count: {len(skewed_features)}")

for feature in skewed_features.index:

    # Ensure positive values
    min_value = all_data[feature].min()

    if min_value < 0:
        all_data[feature] = (
            all_data[feature] + abs(min_value)
        )

    # Apply safe log transform
    all_data[feature] = np.log1p(all_data[feature])

# ============================================================
# PREPROCESSING PIPELINE
# ============================================================

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('cat', categorical_pipeline, categorical_cols)
])

# ============================================================
# SPLIT TRAIN / TEST
# ============================================================

X = all_data.iloc[:len(train_df), :]
X_test = all_data.iloc[len(train_df):, :]

# ============================================================
# MODELS
# ============================================================

xgb_model = XGBRegressor(
    n_estimators=5000,
    learning_rate=0.01,
    max_depth=3,
    min_child_weight=0,
    subsample=0.7,
    colsample_bytree=0.7,
    objective='reg:squarederror',
    reg_alpha=0.0005,
    reg_lambda=0.9,
    random_state=42,
    n_jobs=-1
)

lgb_model = LGBMRegressor(
    objective='regression',
    num_leaves=4,
    learning_rate=0.01,
    n_estimators=5000,
    max_bin=200,
    bagging_fraction=0.75,
    bagging_freq=5,
    feature_fraction=0.2,
    feature_fraction_seed=42,
    bagging_seed=42,
    min_data_in_leaf=3,
    min_sum_hessian_in_leaf=11,
    random_state=42
)

gbr_model = GradientBoostingRegressor(
    n_estimators=3000,
    learning_rate=0.01,
    max_depth=4,
    max_features='sqrt',
    min_samples_leaf=15,
    min_samples_split=10,
    loss='huber',
    random_state=42
)

enet_model = ElasticNet(
    alpha=0.0005,
    l1_ratio=0.9,
    random_state=42
)

# ============================================================
# CREATE PIPELINES
# ============================================================

xgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', xgb_model)
])

lgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', lgb_model)
])

gbr_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', gbr_model)
])

enet_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', enet_model)
])

# ============================================================
# CROSS VALIDATION FUNCTION
# ============================================================

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

def rmsle_cv(model):

    rmse = np.sqrt(
        -cross_val_score(
            model,
            X,
            y,
            scoring='neg_mean_squared_error',
            cv=kf,
            n_jobs=-1
        )
    )

    return rmse

# ============================================================
# MODEL SCORES
# ============================================================

models = {
    'ElasticNet': enet_pipeline,
    'GradientBoosting': gbr_pipeline,
    'XGBoost': xgb_pipeline,
    'LightGBM': lgb_pipeline
}

print("="*70)
print("CROSS VALIDATION RMSE SCORES")
print("="*70)

for name, model in models.items():

    score = rmsle_cv(model)

    print(f"{name}")
    print(f"Mean RMSE : {score.mean():.5f}")
    print(f"Std RMSE  : {score.std():.5f}")
    print("-"*70)

# ============================================================
# TRAIN MODELS
# ============================================================

print("\nTraining Models...\n")

xgb_pipeline.fit(X, y)
lgb_pipeline.fit(X, y)
gbr_pipeline.fit(X, y)
enet_pipeline.fit(X, y)

# ============================================================
# PREDICTIONS
# ============================================================

xgb_pred = np.expm1(xgb_pipeline.predict(X_test))
lgb_pred = np.expm1(lgb_pipeline.predict(X_test))
gbr_pred = np.expm1(gbr_pipeline.predict(X_test))
enet_pred = np.expm1(enet_pipeline.predict(X_test))

# ============================================================
# FINAL BLENDING
# ============================================================

final_predictions = (
    0.35 * xgb_pred +
    0.35 * lgb_pred +
    0.20 * gbr_pred +
    0.10 * enet_pred
)

# ============================================================
# CREATE SUBMISSION FILE
# ============================================================

submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_predictions
})

submission.to_csv('submission.csv', index=False)

print("="*70)
print("SUBMISSION FILE CREATED SUCCESSFULLY")
print("File Name: submission.csv")
print("="*70)

print(submission.head())
```

```python
# ============================================================
# HOUSE PRICES - ADVANCED REGRESSION TECHNIQUES
# FINAL OPTIMIZED KAGGLE SOLUTION
# ============================================================

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from scipy.stats import skew

from sklearn.model_selection import KFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import (
    RobustScaler,
    OneHotEncoder
)

from sklearn.linear_model import ElasticNet

from sklearn.metrics import mean_squared_error

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# ============================================================
# LOAD DATA
# ============================================================

train_df = pd.read_csv(
    '/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv'
)

test_df = pd.read_csv(
    '/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv'
)

test_ids = test_df['Id']

# ============================================================
# REMOVE OUTLIERS
# ============================================================

train_df = train_df.drop(
    train_df[
        (train_df['GrLivArea'] > 4000) &
        (train_df['SalePrice'] < 300000)
    ].index
)

# ============================================================
# TARGET TRANSFORMATION
# ============================================================

y = np.log1p(train_df['SalePrice'])

# ============================================================
# COMBINE DATA
# ============================================================

train_features = train_df.drop(['SalePrice'], axis=1)

all_data = pd.concat(
    [train_features, test_df],
    axis=0
).reset_index(drop=True)

# ============================================================
# MANUAL MISSING VALUE HANDLING
# ============================================================

none_cols = [
    'PoolQC',
    'MiscFeature',
    'Alley',
    'Fence',
    'FireplaceQu',
    'GarageType',
    'GarageFinish',
    'GarageQual',
    'GarageCond',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'MasVnrType'
]

for col in none_cols:
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna('None')

# Numerical missing values
for col in all_data.select_dtypes(include=np.number).columns:
    all_data[col] = all_data[col].fillna(
        all_data[col].median()
    )

# Categorical missing values
for col in all_data.select_dtypes(exclude=np.number).columns:
    all_data[col] = all_data[col].fillna(
        all_data[col].mode()[0]
    )

# ============================================================
# FEATURE ENGINEERING
# ============================================================

def create_features(df):

    df = df.copy()

    # Total square footage
    df['TotalSF'] = (
        df['TotalBsmtSF'] +
        df['1stFlrSF'] +
        df['2ndFlrSF']
    )

    # Total bathrooms
    df['TotalBath'] = (
        df['FullBath'] +
        0.5 * df['HalfBath'] +
        df['BsmtFullBath'] +
        0.5 * df['BsmtHalfBath']
    )

    # Total porch area
    df['TotalPorchSF'] = (
        df['OpenPorchSF'] +
        df['3SsnPorch'] +
        df['EnclosedPorch'] +
        df['ScreenPorch'] +
        df['WoodDeckSF']
    )

    # House age
    df['HouseAge'] = (
        df['YrSold'] - df['YearBuilt']
    )

    # Remodel age
    df['RemodAge'] = (
        df['YrSold'] - df['YearRemodAdd']
    )

    # Binary features
    df['HasGarage'] = (
        df['GarageArea'] > 0
    ).astype(int)

    df['HasBasement'] = (
        df['TotalBsmtSF'] > 0
    ).astype(int)

    df['HasFireplace'] = (
        df['Fireplaces'] > 0
    ).astype(int)

    # Interaction features
    df['OverallQual_TotalSF'] = (
        df['OverallQual'] * df['TotalSF']
    )

    df['OverallQual_GrLivArea'] = (
        df['OverallQual'] * df['GrLivArea']
    )

    df['GarageScore'] = (
        df['GarageArea'] * df['GarageCars']
    )

    df['KitchenScore'] = (
        df['KitchenAbvGr'] * df['OverallQual']
    )

    df['TotalHouseSF'] = (
        df['TotalBsmtSF'] +
        df['1stFlrSF'] +
        df['2ndFlrSF']
    )

    return df

all_data = create_features(all_data)

# ============================================================
# DROP WEAK FEATURES
# ============================================================

drop_cols = ['Id']

all_data.drop(columns=drop_cols, inplace=True)

# ============================================================
# HANDLE SKEWNESS
# ============================================================

numeric_cols = all_data.select_dtypes(
    include=np.number
).columns

skewness = all_data[numeric_cols].apply(
    lambda x: skew(x.dropna())
)

skewed_features = skewness[
    abs(skewness) > 0.75
].index

for col in skewed_features:

    min_value = all_data[col].min()

    if min_value < 0:
        all_data[col] = (
            all_data[col] + abs(min_value)
        )

    all_data[col] = np.log1p(all_data[col])

# ============================================================
# COLUMN TYPES
# ============================================================

numeric_cols = all_data.select_dtypes(
    include=np.number
).columns

categorical_cols = all_data.select_dtypes(
    exclude=np.number
).columns

# ============================================================
# PREPROCESSING
# ============================================================

numeric_pipeline = Pipeline([
    (
        'imputer',
        SimpleImputer(strategy='median')
    ),
    (
        'scaler',
        RobustScaler()
    )
])

categorical_pipeline = Pipeline([
    (
        'imputer',
        SimpleImputer(strategy='most_frequent')
    ),
    (
        'encoder',
        OneHotEncoder(
            handle_unknown='ignore'
        )
    )
])

preprocessor = ColumnTransformer([
    (
        'num',
        numeric_pipeline,
        numeric_cols
    ),
    (
        'cat',
        categorical_pipeline,
        categorical_cols
    )
])

# ============================================================
# SPLIT TRAIN / TEST
# ============================================================

X = all_data.iloc[:len(train_df), :]
X_test = all_data.iloc[len(train_df):, :]

# ============================================================
# MODELS
# ============================================================

xgb_model = XGBRegressor(
    n_estimators=5000,
    learning_rate=0.01,
    max_depth=3,
    min_child_weight=0,
    gamma=0,
    subsample=0.7,
    colsample_bytree=0.7,
    reg_alpha=0.0005,
    reg_lambda=0.9,
    objective='reg:squarederror',
    random_state=42,
    n_jobs=-1
)

lgb_model = LGBMRegressor(
    objective='regression',
    n_estimators=5000,
    learning_rate=0.005,
    num_leaves=4,
    max_depth=-1,
    subsample=0.6,
    colsample_bytree=0.3,
    reg_alpha=0.1,
    reg_lambda=0.1,
    random_state=42
)

cat_model = CatBoostRegressor(
    iterations=5000,
    learning_rate=0.01,
    depth=4,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_seed=42,
    verbose=0
)

enet_model = ElasticNet(
    alpha=0.0005,
    l1_ratio=0.9,
    random_state=42
)

# ============================================================
# PIPELINES
# ============================================================

xgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', xgb_model)
])

lgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', lgb_model)
])

cat_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', cat_model)
])

enet_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', enet_model)
])

# ============================================================
# CROSS VALIDATION
# ============================================================

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

def rmse_cv(model):

    rmse = np.sqrt(
        -cross_val_score(
            model,
            X,
            y,
            scoring='neg_mean_squared_error',
            cv=kf,
            n_jobs=-1
        )
    )

    return rmse

# ============================================================
# MODEL SCORES
# ============================================================

models = {
    'XGBoost': xgb_pipeline,
    'LightGBM': lgb_pipeline,
    'CatBoost': cat_pipeline,
    'ElasticNet': enet_pipeline
}

print("="*70)
print("CROSS VALIDATION RMSE")
print("="*70)

for name, model in models.items():

    score = rmse_cv(model)

    print(f"{name}")
    print(f"Mean RMSE : {score.mean():.5f}")
    print(f"Std RMSE  : {score.std():.5f}")
    print("-"*70)

# ============================================================
# TRAIN MODELS
# ============================================================

print("\nTraining Models...\n")

xgb_pipeline.fit(X, y)
lgb_pipeline.fit(X, y)
cat_pipeline.fit(X, y)
enet_pipeline.fit(X, y)

# ============================================================
# PREDICTIONS
# ============================================================

xgb_pred = np.expm1(
    xgb_pipeline.predict(X_test)
)

lgb_pred = np.expm1(
    lgb_pipeline.predict(X_test)
)

cat_pred = np.expm1(
    cat_pipeline.predict(X_test)
)

enet_pred = np.expm1(
    enet_pipeline.predict(X_test)
)

# ============================================================
# FINAL BLENDING
# ============================================================

final_predictions = (
    0.30 * xgb_pred +
    0.30 * lgb_pred +
    0.30 * cat_pred +
    0.10 * enet_pred
)

# ============================================================
# CREATE SUBMISSION
# ============================================================

submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_predictions
})

submission.to_csv(
    'submission.csv',
    index=False
)

print("="*70)
print("SUBMISSION FILE CREATED SUCCESSFULLY")
print("File Name: submission.csv")
print("="*70)

print(submission.head())
```

```python
# ============================================================
# HOUSE PRICES - ADVANCED REGRESSION TECHNIQUES
# OPTIMIZED KAGGLE SOLUTION
# ============================================================

# ============================================================
# IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from scipy.stats import skew
from scipy.special import boxcox1p
from scipy.stats import boxcox_normmax

from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, OneHotEncoder

from sklearn.linear_model import ElasticNet
from sklearn.ensemble import GradientBoostingRegressor

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# ============================================================
# LOAD DATA
# ============================================================

train_df = pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv')

test_df = pd.read_csv(
    '/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv'
)

# Save IDs
train_ids = train_df['Id']
test_ids = test_df['Id']

# ============================================================
# REMOVE OUTLIERS
# ============================================================

# train_df = train_df.drop(
#     train_df[
#         (train_df['GrLivArea'] > 4000) &
#         (train_df['SalePrice'] < 300000)
#     ].index
# )

# ============================================================
# TARGET TRANSFORMATION
# ============================================================

# Kaggle metric uses log
y = np.log1p(train_df['SalePrice'])

# ============================================================
# COMBINE TRAIN + TEST
# ============================================================

train_features = train_df.drop(['SalePrice'], axis=1)
all_data = pd.concat([train_features, test_df]).reset_index(drop=True)

# ============================================================
# FEATURE ENGINEERING
# ============================================================

def create_features(df):

    df = df.copy()

    # Total SF
    df['TotalSF'] = (
        df['TotalBsmtSF'] +
        df['1stFlrSF'] +
        df['2ndFlrSF']
    )

    # Total bathrooms
    df['TotalBath'] = (
        df['FullBath'] +
        0.5 * df['HalfBath'] +
        df['BsmtFullBath'] +
        0.5 * df['BsmtHalfBath']
    )

    # Total porch area
    df['TotalPorchSF'] = (
        df['OpenPorchSF'] +
        df['3SsnPorch'] +
        df['EnclosedPorch'] +
        df['ScreenPorch'] +
        df['WoodDeckSF']
    )

    # House age
    df['HouseAge'] = df['YrSold'] - df['YearBuilt']

    # Remodel age
    df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']

    # Has basement
    df['HasBsmt'] = (df['TotalBsmtSF'] > 0).astype(int)

    # Has garage
    df['HasGarage'] = (df['GarageArea'] > 0).astype(int)

    # Has fireplace
    df['HasFireplace'] = (df['Fireplaces'] > 0).astype(int)

    return df

all_data = create_features(all_data)

# ============================================================
# DROP VERY HIGH MISSING COLUMNS
# ============================================================

drop_cols = [
    'PoolQC',
    'MiscFeature',
    'Alley',
    'Fence'
]

all_data.drop(columns=drop_cols, inplace=True)

# ============================================================
# NUMERICAL & CATEGORICAL COLUMNS
# ============================================================

numeric_cols = all_data.select_dtypes(include=np.number).columns
categorical_cols = all_data.select_dtypes(exclude=np.number).columns

# ============================================================
# SIMPLE SKEWNESS HANDLING (STABLE)
# ============================================================

from scipy.stats import skew

numeric_cols = all_data.select_dtypes(include=np.number).columns

skewed_features = all_data[numeric_cols].apply(
    lambda x: skew(x.dropna())
)

skewed_features = skewed_features[
    abs(skewed_features) > 1
]

print(f"Skewed Features Count: {len(skewed_features)}")

for feature in skewed_features.index:

    # Ensure positive values
    min_value = all_data[feature].min()

    if min_value < 0:
        all_data[feature] = (
            all_data[feature] + abs(min_value)
        )

    # Apply safe log transform
    all_data[feature] = np.log1p(all_data[feature])

# ============================================================
# PREPROCESSING PIPELINE
# ============================================================

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('cat', categorical_pipeline, categorical_cols)
])

# ============================================================
# SPLIT TRAIN / TEST
# ============================================================

X = all_data.iloc[:len(train_df), :]
X_test = all_data.iloc[len(train_df):, :]

# ============================================================
# MODELS
# ============================================================

xgb_model = XGBRegressor(
    n_estimators=5000,
    learning_rate=0.01,
    max_depth=3,
    min_child_weight=0,
    subsample=0.7,
    colsample_bytree=0.7,
    objective='reg:squarederror',
    reg_alpha=0.0005,
    reg_lambda=0.9,
    random_state=42,
    n_jobs=-1
)

lgb_model = LGBMRegressor(
    objective='regression',
    num_leaves=4,
    learning_rate=0.01,
    n_estimators=5000,
    max_bin=200,
    bagging_fraction=0.75,
    bagging_freq=5,
    feature_fraction=0.2,
    feature_fraction_seed=42,
    bagging_seed=42,
    min_data_in_leaf=3,
    min_sum_hessian_in_leaf=11,
    random_state=42
)

gbr_model = GradientBoostingRegressor(
    n_estimators=3000,
    learning_rate=0.01,
    max_depth=4,
    max_features='sqrt',
    min_samples_leaf=15,
    min_samples_split=10,
    loss='huber',
    random_state=42
)

enet_model = ElasticNet(
    alpha=0.0005,
    l1_ratio=0.9,
    random_state=42
)

# ============================================================
# CREATE PIPELINES
# ============================================================

xgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', xgb_model)
])

lgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', lgb_model)
])

gbr_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', gbr_model)
])

enet_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', enet_model)
])

# ============================================================
# CROSS VALIDATION FUNCTION
# ============================================================

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

def rmsle_cv(model):

    rmse = np.sqrt(
        -cross_val_score(
            model,
            X,
            y,
            scoring='neg_mean_squared_error',
            cv=kf,
            n_jobs=-1
        )
    )

    return rmse

# ============================================================
# MODEL SCORES
# ============================================================

models = {
    'ElasticNet': enet_pipeline,
    'GradientBoosting': gbr_pipeline,
    'XGBoost': xgb_pipeline,
    'LightGBM': lgb_pipeline
}

print("="*70)
print("CROSS VALIDATION RMSE SCORES")
print("="*70)

for name, model in models.items():

    score = rmsle_cv(model)

    print(f"{name}")
    print(f"Mean RMSE : {score.mean():.5f}")
    print(f"Std RMSE  : {score.std():.5f}")
    print("-"*70)

# ============================================================
# TRAIN MODELS
# ============================================================

print("\nTraining Models...\n")

xgb_pipeline.fit(X, y)
lgb_pipeline.fit(X, y)
gbr_pipeline.fit(X, y)
enet_pipeline.fit(X, y)

# ============================================================
# PREDICTIONS
# ============================================================

xgb_pred = np.expm1(xgb_pipeline.predict(X_test))
lgb_pred = np.expm1(lgb_pipeline.predict(X_test))
gbr_pred = np.expm1(gbr_pipeline.predict(X_test))
enet_pred = np.expm1(enet_pipeline.predict(X_test))

# ============================================================
# FINAL BLENDING
# ============================================================

final_predictions = (
    0.35 * xgb_pred +
    0.35 * lgb_pred +
    0.20 * gbr_pred +
    0.10 * enet_pred
)

# ============================================================
# CREATE SUBMISSION FILE
# ============================================================

submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_predictions
})

submission.to_csv('submission.csv', index=False)

print("="*70)
print("SUBMISSION FILE CREATED SUCCESSFULLY")
print("File Name: submission.csv")
print("="*70)

print(submission.head())
```
